### Vaccination in India

```python
vaccine = pd.read_csv('covid_vaccine_statewise.csv')
```

```python
vaccine.head(5)
```

```python
vaccine.tail()
```

```python
vaccine.rename(columns = {'Updated On':'Date'}, inplace = True)
```

```python
vaccine.tail()
```

```python
vaccine.shape
```

```python
vaccine.info()
```

```python
def autopct(pct): # only show the label when it's > 10%
    if pct < 1:
        return ('%.3f' % pct)
    else:
        return ('%.1f' % pct)

male = vaccine["Male(Individuals Vaccinated)"].sum()/vaccine["Total Individuals Vaccinated"].sum()
female = vaccine["Female(Individuals Vaccinated)"].sum()/vaccine["Total Individuals Vaccinated"].sum()
trans = vaccine["Transgender(Individuals Vaccinated)"].sum()/vaccine["Total Individuals Vaccinated"].sum()
vaccine_ratio = [male,female,trans]
mylabels = ["Male", "Female", "Transgender"]
plt.figure(figsize=(15, 8))
plt.pie(vaccine_ratio, labels=mylabels, autopct=autopct)
plt.title("Vaccination Ratio till 15/05/21")
plt.legend()
plt.show()
```

```python
covishield = vaccine['Total CoviShield Administered'].sum()/vaccine["Total Doses Administered"].sum()
covaxin = vaccine['Total Covaxin Administered'].sum()/vaccine["Total Doses Administered"].sum()
vaccineType_ratio = [covishield,covaxin]
mylabels = ["CoviShield", "Covaxin"]
colors = ['gold', 'lightskyblue']
plt.pie(vaccineType_ratio, labels=mylabels, autopct=autopct)
plt.title("Type of Vaccine administered till 15/05/21")
plt.get_cmap('bwr')
plt.legend()
plt.show()
```
